```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')

```

# Task 1 - Normalize Data in all columns

- *Load the Network dataset*
- *Define function to normalize columns using Min-Max Normalization*
- *Normalize network features and visualize refactored data*
- *Save normalized dataset*

```python
# Normalizing data from all the columns

df = pd.read_csv('./data/github-full-dataset.csv')
df = df.drop(columns=['Unnamed: 0'])
df.head(10) # Current Dataset before Normalization
```

```python
# Function to scale a DataFrame column to a range of 0 to 1 (i.e., Normalization)
def normalize_column(data_series):
    return (data_series - data_series.min()) / (data_series.max() - data_series.min())
```

```python
# Define a function to create detailed histograms of our normalized data

def plot_detailed_histogram(input_data, plot_title, x_axis_label, y_axis_label='Frequency', plot_color='skyblue'):
    plt_edge_color = "black"
    plt.figure(figsize=(10, 6))  
    plt.hist(input_data, bins=50, color=plot_color, edgecolor=plt_edge_color)  
    plt.title(plot_title)  
    plt.xlabel(x_axis_label)  
    plt.ylabel(y_axis_label)  
    plt.grid(True, linestyle='--', alpha=0.6)  
    plt.show()
```

```python
# Normalizing Page_Rank_Source

df['Page_Rank_Source'] = normalize_column(df['Page_Rank_Source'])
plot_detailed_histogram(df['Page_Rank_Source'], 'Distribution of Page Rank (Source)', 'Page Rank (Source)', plot_color='teal')
```

```python
# Normalizing Page_Rank_Target

df['Page_Rank_Target'] = normalize_column(df['Page_Rank_Target'])
plot_detailed_histogram(df['Page_Rank_Target'], 'Distribution of Page Rank (Target)', 'Page Rank (Target)', plot_color='purple')
```

```python
# Normalizing Shortest Path

df['Shortest_Path'] = normalize_column(df['Shortest_Path'])
plot_detailed_histogram(df['Shortest_Path'], 'Distribution of Shortest Path', 'Shortest Path', plot_color='red')
```

```python
# Normalizing Source_Followers

df['Source_Followers'] = normalize_column(df['Source_Followers'])
plot_detailed_histogram(df['Source_Followers'], 'Distribution of Source Followers', 'Source Followers', plot_color='blue')
```

```python
# Normalizing Source_Following

df['Source_Following'] = normalize_column(df['Source_Following'])
plot_detailed_histogram(df['Source_Following'], 'Distribution of Source Following', 'Source Following', plot_color='green')
```

```python
# Normalizing Target_Followers

df['Target_Followers'] = normalize_column(df['Target_Followers'])
plot_detailed_histogram(df['Target_Followers'], 'Distribution of Target Followers', 'Target Followers', plot_color='orange')
```

```python
# Normalizing Target_Following

df['Target_Following'] = normalize_column(df['Target_Following'])
plot_detailed_histogram(df['Target_Following'], 'Distribution of Target Following', 'Target Following', plot_color='magenta')
```

```python
# Normalizing Mutual_Followers

df['Mutual_Followers'] = normalize_column(df['Mutual_Followers'])
plot_detailed_histogram(df['Mutual_Followers'], 'Distribution of Mutual Followers', 'Mutual Followers', plot_color='brown')
```

```python
# Normalizing Mutual_Following

df['Mutual_Following'] = normalize_column(df['Mutual_Following'])
plot_detailed_histogram(df['Mutual_Following'], 'Distribution of Mutual Following', 'Mutual Following', plot_color='grey')
```
